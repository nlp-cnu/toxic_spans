```python
import os

from ast import literal_eval
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
metrics_path = os.path.join('toxic_spans_full_model', 'metrics.csv')
met_df = pd.read_csv(metrics_path)

for key in met_df.keys():
    if key != 'epoch':
        met_df[key] = [eval(met_df[key][i]) for i, x in enumerate(met_df[key])]


loss = list(zip(*met_df['loss']))
f1 = list(zip(*met_df['f1']))
precision = list(zip(*met_df['precision']))
recall = list(zip(*met_df['recall']))
accuracy = list(zip(*met_df['accuracy']))

train_loss = list(loss[0])
eval_loss = list(loss[1])
train_f1 = list(f1[0])
eval_f1 = list(f1[1])
train_precision = list(precision[0])
eval_precision = list(precision[1])
train_recall = list(recall[0])
eval_recall = list(recall[1])
train_accuracy = list(accuracy[0])
eval_accuracy = list(accuracy[1])

epochs = np.arange(1, len(met_df['loss']) + 1)

figure1, ax1 = plt.subplots(figsize=[15,10])
# ax1.plot(epochs, train_loss, label='Train Loss')
ax1.plot(epochs, train_f1, label='Machine Learning F1 score', linewidth = 4)
ax1.plot(epochs, train_precision, label='Machine Learning Precision', linewidth = 4)
ax1.plot(epochs, [(x / 2) for x in train_recall], label='Recall', linewidth = 4)
# ax1.plot(epochs, train_accuracy, label='Accuracy')
# ax1.plot(epochs, [.45138]*len(epochs), 'o', label='Dictionary 5F-CV F1 score')

handles, labels = ax1.get_legend_handles_labels()
plt.legend(handles, labels, loc='lower right')

ax1.set_ylim((0, 1))
ax1.set_title('Training metrics')
ax1.set_xlabel('Number of epochs')
ax1.set_ylabel('Scores')

figure2, ax2 = plt.subplots(figsize=[15,10])
# ax2.plot(epochs, eval_loss, label='Train Loss', linewidth = 4)
ax2.plot(epochs, eval_f1, label='Machine Learning F1 score', linewidth = 4)
ax2.plot(epochs, eval_precision, label='Machine Learning Precision Score', linewidth = 4)
ax2.plot(epochs, [(x / 2) for x in eval_recall], label='Machine Learning Recall Score', linewidth = 4)
# ax2.plot(epochs, eval_accuracy, label='Accuracy')
# ax2.plot(epochs, [.45138]*len(epochs), 'o', label='Dictionary 5F-CV F1 score')
ax2.plot(epochs, [.45138]*len(epochs), 'o', label='Dictionary F1 score')

handles, labels = ax2.get_legend_handles_labels()
plt.legend(handles, labels, loc='lower right')

ax2.set_ylim((0, 1))
ax2.set_title('Test set metrics')
ax2.set_xlabel('Number of epochs')
ax2.set_ylabel('Scores')

SMALL_SIZE = 10
MEDIUM_SIZE = 15
BIGGER_SIZE = 25

plt.rc('font', size=28)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)

plt.show()
```
